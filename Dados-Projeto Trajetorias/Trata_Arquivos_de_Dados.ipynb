```python
import pandas as pd
import openpyxl
```

Selecionando os dados de malária:

```python
epidem_indic = pd.read_excel('TRAJETORIAS_DATASET_Epidemiological_dimension_indicators.xlsx')

# Define os valores desejados
malaria = ['Falciparum', 'Vivax', 'Vivax+Falciparum']

# Filtra as linhas com os valores desejados na coluna 'disease'
malaria_df = epidem_indic[epidem_indic['disease'].isin(malaria)]

# Exibe as linhas filtradas
malaria_df = malaria_df.reset_index()
malaria_df = malaria_df.drop(['index'], axis=1)
malaria_df
```

```python
malaria_df.to_excel('malaria_dataset.xlsx', index=False)
```

Buscando o estado com maior número de casos totais:

```python
periodos_desejados = ['2004-2008', '2015-2019']
filtros = (epidem_indic['disease'].isin(malaria)) & (epidem_indic['period'].isin(periodos_desejados))
dados_filtrados = epidem_indic[filtros]

# Agrupa os dados por estado e soma os casos
soma_casos_por_estado = dados_filtrados.groupby('state')['cases'].sum()

# Ordena os estados em ordem decrescente de casos
soma_casos_por_estado = soma_casos_por_estado.sort_values(ascending=False)

# Obtém o estado com o maior número de casos
estado_maior_casos = soma_casos_por_estado.index[0]

#############################

# Agrupa os dados por estado e período

soma_casos_por_estado_periodo = dados_filtrados.groupby(['state', 'period'])['cases'].sum()

soma_casos_por_estado_periodo = soma_casos_por_estado_periodo.sort_values(ascending=False)

estado_periodo_maior_casos = soma_casos_por_estado_periodo.index[0]


# Exibe o estado com o maior índice de casos
print(soma_casos_por_estado)
print(' ')
print("Estado com maior índice de casos:", estado_maior_casos)
```

Buscando o estado com maior número de casos por período:

```python
# Exibe o estado com o maior índice de casos
print(soma_casos_por_estado_periodo)
print(' ')
print("Estado com maior índice de casos:", estado_periodo_maior_casos)
```
